```python
PATH_TO_DATASET ="/home/basile/Documents/projet_bees_detection_basile/bees_detection/src/datafiles/final_datafiles/dataset_yolo_cropped_with_cleaned_structure.csv"

df = pd.read_csv(PATH_TO_DATASET)


h = np.zeros(20000)
w = np.zeros(20000)


# take 200 random images
for i in range(20000):

    rand = np.random.randint(0, len(df['Paths']))
    img = cv2.imread(df['Paths'][rand])
    h[i] = img.shape[0]
    w[i] = img.shape[1]




print("Average height: ", np.mean(h))
print("Average width: ", np.mean(w))


print("Max height: ", np.max(h))
print("Max width: ", np.max(w))


print("Min height: ", np.min(h))
print("Min width: ", np.min(w))


print("Std height: ", np.std(h))
print("Std width: ", np.std(w))

```

```python
# plot the distribution of the height and width
import matplotlib.pyplot as plt


# keep only the images with a height < 1000
h = h[h<1000]
w = w[w<1000]


plt.figure(figsize=(10,10))

plt.subplot(2,1,1)
plt.hist(h, bins=100)
plt.title("Distribution of the height of the images in the dataset")

plt.subplot(2,1,2)
plt.hist(w, bins=100)
plt.title("Distribution of the width of the images in the dataset")


plt.show()
```
