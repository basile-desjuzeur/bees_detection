# Recreate a proper directory structure

The goal is to recreate the following hierarchy : 

path/to/directory/label/img.jpg

```python
path = '/workspaces/projet_bees_detection_basile/data_bees_detection/Nouvelles photos après 21avril'

new_path = '/workspaces/projet_bees_detection_basile/data_bees_detection/new_justine_post_21_avril'

os.makedirs(new_path, exist_ok=True)


for root,dir,file in os.walk(path):

    for f in file: 

        # dir path
        dir_path = root.split('/')[-2]
        if not os.path.exists(os.path.join(new_path,dir_path)):
            os.makedirs(os.path.join(new_path,dir_path), exist_ok=True)

        cmd = 'cp "' + os.path.join(root,f) + '" "' + os.path.join(new_path,dir_path) + '"'
        os.system(cmd)

        print(cmd)
```
